```python
def optimization_DC1201_rf(Total_flow_EL_lower,Total_flow_EL_upper,
HCl_per_EL_in_Feed_lower,HCl_per_EL_in_Feed_upper,
O2_per_EL_in_Feed_lower,O2_per_EL_in_Feed_upper,
Rec_per_EL_in_feed_lower,Rec_per_EL_in_feed_upper,
Vol_EL_in_RG_lower,Vol_EL_in_RG_upper,
Temp_C_lower,Temp_C_upper,
P_OVH_lower,P_OVH_upper,
Cat_inventory_ton_lower,Cat_inventory_ton_upper,
init_points=5,n_iter=100):
    """
    Perform Bayesian Optimization to find optimal DC-1201 reactor conditions
    that maximize EDC purity by minimizing key impurities (2-chloro, 112-TCE, CCl4).

    Parameters
    ----------
    For each variable below, specify a lower and upper bound.
    If optimization is not required for a variable, set both bounds equal.

    Total_flow_EL : float
        Ethylene feed flow rate (Ncum/hr)
    HCl_per_EL_in_Feed : float
        Volumetric ratio of HCl per ethylene
    O2_per_EL_in_Feed : float
        Volumetric ratio of O2 per ethylene
    Rec_per_EL_in_feed : float
        Volumetric ratio of recycle gas per ethylene feed
    Vol_EL_in_RG : float
        Volumetric flow rate of ethylene in recycle gas (Ncum/hr)
    Temp_C : float
        Reactor temperature (°C)
    P_OVH : float
        Reactor pressure (kg/cm²)
    Cat_inventory_ton : float
        Catalyst inventory (tons)
    init_points : int, optional (default=5)
        Number of initial random exploration steps
    n_iter : int, optional (default=100)
        Number of Bayesian Optimization steps

    Returns
    -------
    pd.DataFrame
        Predicted optimal component composition:
        - Purity EDC (%wt)
        - 2-chloro (%ww)
        - 112 TCE (%ww)
        - CCl4 (%ww)

    Notes
    -----
    This function uses a trained Random Forest model ('DC1201_rf_model_2.sav')
    to predict component composition. The optimization objective is:
        Purity EDC - 2-chloro - 112 TCE - CCl4
    """
    from bayes_opt import BayesianOptimization
    import pickle
    pbounds = {
    "Total_flow_EL": (Total_flow_EL_lower, Total_flow_EL_upper),
    "HCl_EL_in_Feed": (HCl_per_EL_in_Feed_lower,HCl_per_EL_in_Feed_upper),
    "O2_EL_in_Feed": (O2_per_EL_in_Feed_lower, O2_per_EL_in_Feed_upper),
    "Rec_EL_in_feed": (Rec_per_EL_in_feed_lower, Rec_per_EL_in_feed_upper),
    "Vol_EL_in_RG": (Vol_EL_in_RG_lower, Vol_EL_in_RG_upper),
    "Temp_C": (Temp_C_lower, Temp_C_upper),
    "P_OVH": (P_OVH_lower, P_OVH_upper),
    "Cat_inventory_ton": (Cat_inventory_ton_lower, Cat_inventory_ton_upper),
}
    def predict_purity(Total_flow_EL, HCl_EL_in_Feed, O2_EL_in_Feed, Rec_EL_in_feed,
                    Vol_EL_in_RG, Temp_C, P_OVH, Cat_inventory_ton):
        
        input_data = pd.DataFrame([[Total_flow_EL, HCl_EL_in_Feed, O2_EL_in_Feed, Rec_EL_in_feed,
                                    Vol_EL_in_RG, Temp_C, P_OVH, Cat_inventory_ton]],
                                columns=['Total flow EL (Ncum/hr)', 'HCl / EL in Feed', 'O2 / El in Feed',
                                        'Rec / El in feed', 'volume El in RG (Ncum/hr)', 
                                        'Temp (C)', 'P.OVH', 'cat. inventory (tone)'])
        rf = pickle.load(open('DC1201_rf_model_2.sav','rb'))
        prediction = rf.predict(input_data)[0]
        return prediction[0] - prediction[1] - prediction[2] - prediction[3]
    optimizer = BayesianOptimization(
        f=predict_purity,
        pbounds=pbounds,
        random_state=42,
    )

    optimizer.maximize(
        init_points=5,
        n_iter=100
    )

    print("Best input conditions:", optimizer.max['params'])
    print("Predicted maximized EDC - (impurities):", optimizer.max['target'])
    condition = optimizer.max['params']
    optimal_condtition = pd.DataFrame([condition])
    optimal_condtition = optimal_condtition.rename(columns={'Cat_inventory_ton':'cat. inventory (tone)',
                                                            'HCl_EL_in_Feed' : 'HCl / EL in Feed',
                                                            'O2_EL_in_Feed'  : 'O2 / El in Feed',
                                                            'Rec_EL_in_feed' :'Rec / El in feed',
                                                            'Temp_C' :'Temp (C)',
                                                            'P_OVH' : 'P.OVH',
                                                            'Total_flow_EL' :'Total flow EL (Ncum/hr)',
                                                            'Vol_EL_in_RG':'volume El in RG (Ncum/hr)'})
    optimal_condtition = optimal_condtition[['Total flow EL (Ncum/hr)', 'HCl / EL in Feed', 'O2 / El in Feed',
        'Rec / El in feed', 'volume El in RG (Ncum/hr)', 'Temp (C)', 'P.OVH',
        'cat. inventory (tone)']]
    rf = pickle.load(open('DC1201_rf_model_2.sav','rb'))
    components = rf.predict(optimal_condtition)
    com = pd.DataFrame(components,columns=['Purity EDC (%wt)', '2-chloro (%ww)', '112 TCE (%ww)', 'CCl4 (%ww)'])
    
    print(com)
    return com
```
